```python
def sigmoid(z):
    return 1 / (1 + np.e ** -z)
```

```python
x = np.arange(-10, 10)
plt.plot(x, sigmoid(x))
plt.grid()
```

```python
def cost_liniar_reg(y, _y):
    return np.mean((y  - _y) ** 2)


def cost_logistic_reg(y, _y):
    return -np.mean(y * np.log(_y) + (1 - y) * np.log(1 - _y))


def h(x, w0, w1=1):
    return sigmoid(w1 * x + w0)
```

```python
arr = []
brr = []

for w0 in np.linspace(-10, 20, 20):
    res = h(x, -w0)
    arr.append(cost_logistic_reg(y, res))
    brr.append(cost_liniar_reg(y, res))
```

```python
plt.plot(arr)
plt.plot(brr)
```
