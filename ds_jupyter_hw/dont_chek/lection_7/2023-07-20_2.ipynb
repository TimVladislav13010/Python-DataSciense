```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/kaggle/input/spotify-recommendation/data.csv")
```

```python
df = df.drop(["time_signature", "liked"], axis=1)
```

```python
df
```

```python
norm_df = (df - df.min()) / (df.max() - df.min())
```

```python
norm_df
```

```python
sample = {
    'danceability': 0.561,
 'energy': 0.9149,
 'key': 1.0,
 'loudness': 50.528,
 'mode': 1.0,
 'speechiness': 0.0419,
 'acousticness': 0.992,
 'instrumentalness': 0.897,
 'liveness': 0.902,
 'valence': 0.6382,
 'tempo': 75.296,
 'duration_ms': 286987.0}
```

```python
sample_series = pd.Series(sample) 
```

```python
norm_sample_series = (sample_series - df.min()) / (df.max() - df.min())
```

```python
recs = np.sqrt(np.sum((norm_df - norm_sample_series) ** 2, axis=1))
```
